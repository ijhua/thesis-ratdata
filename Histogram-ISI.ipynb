```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
%matplotlib inline
```

```python
data = pd.read_csv('DD11_forhist_ISI.csv', sep=',',header=0, index_col =0)

n_bins = 30

x = len(data)
y = data["ISI1"]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(y,n_bins)
plt.title("DD08 Channel 1")
plt.xlabel("ISI")
plt.ylabel("Count")
```
