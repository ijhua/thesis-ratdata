```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#frequency channel 1
files = ["DD11_forhist.csv", "DD08_forhist.csv"]

for file in files:
    data = pd.read_csv(file, sep=',',header=0, index_col =0)
    subject = file.split("_")[0]
    n_bins = 50
    x = len(data)
    y = data["Frequency1"]
    fig = plt.figure(figsize = (15,5))
    ax = fig.add_subplot(111)
    ax.hist(y,n_bins)
    plt.title(subject + " Channel 1 Frequency Histogram")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Count")
    ax.set_xlim([0,60])
    ax.set_ylim([0,1400])
    fig.savefig("Figures/" + subject + "_frequency_histogram_ch1.png")
```

```python
# ISI channel 1
files = ["DD11_forhist_ISI.csv", "DD08_forhist_ISI.csv"]

for file in files:
    data = pd.read_csv(file, sep=',',header=0, index_col =0)
    subject = file.split("_")[0]

    #keep values between 0 and 0.5
    data = data.mask(data.ISI1>0.5)

    n_bins = 100

    x = len(data)
    y = data["ISI1"]
    fig = plt.figure(figsize=(15,5))
    ax = fig.add_subplot(111)

    ax.hist(y,n_bins)
    plt.title(subject + " Channel 1 ISI Histogram")
    plt.xlabel("ISI (s)")
    plt.ylabel("Count")
    ax.set_xlim([0,0.6])
    ax.set_ylim([0,300000])
    fig.savefig("Figures/" + subject + "_ISI_histogram_ch1.png")
```
